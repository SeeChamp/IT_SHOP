```python
import pandas as pd

name = ['คีบอร์ด','เมาส์','ตุ๊กตา']
category = ['อุปกรณ์คอมพิวเตอร์','อุปกรณ์คอมพิวเตอร์','ของเล่น']
price = [1000, 500, 300]
```

```python
#series
name_series = pd.Series(name)
category_series = pd.Series(category)
price_series = pd.Series(price)
```

```python
# Create a DataFrame from the lists
frame = {'ชื่อสินค้า': name_series,
         'หมวดหมู่': category_series,  
         'ราคา': price_series}
```

```python
df = pd.DataFrame(frame)
```

```python
df.set_index('ชื่อสินค้า', inplace=True)
```
